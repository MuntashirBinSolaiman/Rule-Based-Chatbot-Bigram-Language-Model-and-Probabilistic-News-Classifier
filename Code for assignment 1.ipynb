````python
import re  # Import the regex module for pattern matching

class ProblemOne:
    # Class-level variable to track the chat status
    is_chat_active = True

    def __init__(self):
        # Instance variable to store the user's message
        self.message_from_user = ""

        # Pattern to detect the word "bug" anywhere in the message
        # This will match "bug", "debug", "buggy", etc.
        self.pattern_for_debug = r'bug'

        # Pattern to detect error-related words: error, crash, or won't compile
        self.pattern_for_error = r'\b(error|crash|won\'t compile)\b'

        # Pattern to detect common confusing programming topics
        self.pattern_for_concepts = r'\b(recursion|loops|concept)\b'

        # Pattern to detect farewell or exit-related phrases
        self.pattern_for_goodbye = r'\b(bye|exit|logout|see you|goodbye)\b'

    def prompt_user(self):
        # Continuously prompt the user for input while the chat is active
        while self.is_chat_active:
            self.message_from_user = input("Student: ").strip()  # Get user input and strip extra spaces
            print(f"User input: {self.message_from_user}")  
            self.codepal_response(self.message_from_user)  

    def codepal_response(self, message):
        # Check if the message mentions "bug" in any form
        if re.search(self.pattern_for_debug, message, re.IGNORECASE):
            print("🤖 CodePal: Debugging can be frustrating. Have you tried printing intermediate results?")

        # Check if the message mentions an error or crash
        elif re.search(self.pattern_for_error, message, re.IGNORECASE):
            print("🤖 CodePal: Let’s take a deep breath and go step-by-step. What does the error message say?")

        # Check if the message contains a confusing concept
        elif re.search(self.pattern_for_concepts, message, re.IGNORECASE):
            print("🤖 CodePal: That’s a common hurdle. Would you like a resource that explains it visually?")

        # Check for farewell or exit phrases
        elif re.search(self.pattern_for_goodbye, message, re.IGNORECASE):
            print("🤖 CodePal: Happy coding! Remember, every bug is a step forward.")
            self.is_chat_active = False  # Stop the chat loop

        # Default response if none of the above patterns matched
        else:
            print("🤖 CodePal: Let’s tackle it together — what’s your next coding challenge?")
            
            
````

# **Problem 2 Author attribution using the bigram language model**

## 📝**Problem Description**
Authorship Attribution is the problem of identifying the author of a given document by looking at other writings by the same author.

For example, given two novels by William Shakespeare and two by Jane Austen, try to identity the potential author of **a sentence** that didn't appear in any of the novels given.

One way of tackling this problem is to learn two bigram models, one for Shakespeare and one Austen. Then lookup the probablities of all the bigrams in the new sentence and multiply them to end up with the total liklihood of the sentence. More concretely:
### Training
1.
$$novel\_by\_austen =  \{gram^{austen}_0, gram^{austen}_1,\dots ,gram^{austen}_N\}$$
$$P(gram_i|austen) = \frac{Count(gram^{austen}_i) + 1}{N + V}$$
$$N : \text{Total number of bigrams in the text by Austen}, V : \text{Number of unique bigrams in the text by Austen} $$  

2.
$$novel\_by\_shakespeare =  \{gram^{shakespeare}_0, gram^{shakespeare}_1,\dots ,gram^{shakespeare}_N\}$$
$$P(gram_i|shakespeare) = \frac{Count(gram^{shakespeare}_i) + 1}{N + V}$$
$$N : \text{Total number of N-grams in the text by Shakespeare}, V : \text{Number of unique n-grams in the text by Shakespeare} $$  

### Testing
$$sentence = \{gram_0, gram_1,\dots ,gram_N\}$$

$$P(sentence | shakespeare)\approx\prod_i P(gram_i|shakespeare)$$
$$P(sentence | austen)\approx\prod_i P(gram_i|austen)$$

$$P(author|sentence) = \max(P(sentence | shakespeare), P(sentence | austen))$$

> **Reminder:** Multiplying many small probabilities can quickly lead to numerical underflow, where the resulting product becomes so small that it is rounded down to zero by the computer. This makes it difficult to distinguish between rare and impossible events.
To avoid this, compute in log space by summing log-probabilities instead of multiplying raw probabilities—this ensures numerical stability and preserves meaningful comparisons between sequences. [Ref link](https://web.stanford.edu/~jurafsky/slp3/slides/lm24aug.pdf)

````python
from nltk.tokenize import sent_tokenize, word_tokenize
import math

# This method splits a big chunk of text into separate sentences.
# Then, it turns each sentence into a list of words.
# So in the end, we get a list of lists – each one is a sentence made of tokens (words).
def tokenize_text_into_sentences(text):
    sentences = sent_tokenize(text)  # Break text into sentences
    tokenized_sentences = []

    for sentence in sentences:
        tokenized_sentences.append(word_tokenize(sentence))  # Turn each sentence into words

    return tokenized_sentences


# This method counts how many times each word (unigram) appears,
# and also counts pairs of words that come one after the other (bigrams).
# It works on a list of tokenized sentences.
def count_ngrams(tokenized_sentences):
    unigram_counts = {}
    bigram_counts = {}

    for sentence in tokenized_sentences:
        for index in range(len(sentence)):
            word = sentence[index]

            # Count single words
            if word in unigram_counts:
                unigram_counts[word] += 1
            else:
                unigram_counts[word] = 1

            # Count word pairs
            if index > 0:
                previous_word = sentence[index - 1]
                bigram = (previous_word, word)

                if bigram in bigram_counts:
                    bigram_counts[bigram] += 1
                else:
                    bigram_counts[bigram] = 1

    return unigram_counts, bigram_counts


# This method takes a sentence and works out how likely it is using a bigram model.
# It uses the counts we got earlier and applies Laplace smoothing to avoid zero probabilities.
# We return the total log probability, which gives us a score for how likely the sentence is.
def calculate_sentence_probability(sentence_tokens, unigram_counts, bigram_counts, vocab_size):
    total_log_probability = 0  # We add logs of probabilities, so we start at 0

    # Go through each word pair in the sentence
    for first_word, second_word in zip(sentence_tokens[:-1], sentence_tokens[1:]):
        bigram = (first_word, second_word)

        # Get how often this pair showed up before
        bigram_count = bigram_counts.get(bigram, 0)

        # Get how often the first word showed up
        unigram_count = unigram_counts.get(first_word, 0)

        # Laplace smoothing: add 1 to numerator and vocab size to denominator
        probability = (bigram_count + 1) / (unigram_count + vocab_size)

        # Add the log of the probability to the total
        total_log_probability += math.log(probability)

    return total_log_probability
````

````python
# We use the method to break both texts into sentences and then into words.
# We do this so we can count how often words and word pairs appear.
sh_tokenized_sentences = tokenize_text_into_sentences(sh_proc)
au_tokenized_sentences = tokenize_text_into_sentences(au_proc)

# Count how many times each word (unigram) and word pair (bigram) shows up.
# This gives us the basic data we need to calculate probabilities later.
sh_unigrams, sh_bigrams = count_ngrams(sh_tokenized_sentences)
au_unigrams, au_bigrams = count_ngrams(au_tokenized_sentences)

# Find out how many unique words are in each author's text.
# We’ll use this number when smoothing the probabilities.
sh_vocab_size = len(set([word for sentence in sh_tokenized_sentences for word in sentence]))
au_vocab_size = len(set([word for sentence in au_tokenized_sentences for word in sentence]))

# Here we write the test sentence we want to check.
# This one is in the style of Shakespeare.
test_sentence_sh = "when shall we three meet againe in thunder lightning or in raine"
test_token_sh = word_tokenize(test_sentence_sh)

# This is the second test sentence, which is from Jane Austen’s style.
test_sentence_au = "elinor this eldest daughter whose advice was so effectual possessed a strength of understanding and coolness of judgment which qualified her though only nineteen to be the counsellor of her mother and enabled her frequently to counteract to the advantage of them all that eagerness of mind in mrs"
test_token_au = word_tokenize(test_sentence_au)

# Now we calculate how likely each sentence is, based on each author's writing.
# We do this using the method we wrote earlier that works with bigram probabilities.
shakespeare_prob_first_sentence = calculate_sentence_probability(test_token_sh, sh_unigrams, sh_bigrams, sh_vocab_size)
austen_prob_first_sentence = calculate_sentence_probability(test_token_sh, au_unigrams, au_bigrams, au_vocab_size)

shakespeare_prob_second_sentence = calculate_sentence_probability(test_token_au, sh_unigrams, sh_bigrams, sh_vocab_size)
austen_prob_second_sentence = calculate_sentence_probability(test_token_au, au_unigrams, au_bigrams, au_vocab_size)

# Check which author is more likely to have written the first test sentence.
# We compare the probability scores and choose the higher one.
if shakespeare_prob_first_sentence > austen_prob_first_sentence:
    most_likely_author_first_sentence = "Shakespeare"
else:
    most_likely_author_first_sentence = "Austen"

# Print the results for the first test sentence
print("Test Sentence 1 =", test_sentence_sh)
print("Probability of the sentence by Shakespeare:", shakespeare_prob_first_sentence)
print("Probability of the sentence by Austen:", austen_prob_first_sentence)
print("The most likely author for the first sentence is:", most_likely_author_first_sentence)
print()

# Do the same thing for the second test sentence
# Again, we compare which author’s style the sentence fits better
if shakespeare_prob_second_sentence > austen_prob_second_sentence:
    most_likely_author_second_sentence = "Shakespeare"
else:
    most_likely_author_second_sentence = "Austen"

# Print the results for the second test sentence
print("Test Sentence 2 =", test_sentence_au)
print("Probability of the sentence by Shakespeare:", shakespeare_prob_second_sentence)
print("Probability of the sentence by Austen:", austen_prob_second_sentence)
print("The most likely author for the second sentence is:", most_likely_author_second_sentence)
````

````python
# Step 3: Compute Bag-of-Words using NLTK

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# This method tokenizes the input text and removes stopwords and non-alphabetical characters.
# It helps in preparing clean data for further natural language processing tasks like text classification.

def tokenize(text):
    # Convert all text to lowercase so words like "The" and "the" are treated the same
    # This ensures consistency and avoids treating the same word differently due to capitalization.
    text = text.lower()

    # Break the text into individual words (tokens)
    # Tokenizing splits the input text into a list of words for easier processing.
    tokens = word_tokenize(text)

    # Get the list of common stopwords in English
    # Stopwords like 'the', 'is', and 'in' are not meaningful for most NLP tasks, so they will be removed.
    stop_words = stopwords.words('english')

    # Keep only the words that are alphabetic and not in the stopwords list
    # This loop filters out non-alphabetic tokens and stopwords, leaving only meaningful words for analysis.
    words = []
    for word in tokens:
        if word.isalpha() and word not in stop_words:
            words.append(word)

    # Return the cleaned list of words
    # The method returns a list of meaningful words, which can be used in further NLP tasks like word frequency analysis.
    return words

def create_frequency_dictionary(data, labels):
    # Initialize an empty dictionary to store the frequency count
    frequency_dictionary = {}
    
    # Iterate through each text and its corresponding label in the dataset
    for text, label in zip(data, labels):
        # Tokenize the text to split it into individual words
        words = tokenize(text)
        
        # For each word in the tokenized text, update the frequency count for the associated label ('FAKE' or 'REAL')
        for word in words:
            # Check if the word is already in the dictionary
            if word not in frequency_dictionary:
                # If it's not, initialize it with default values {'FAKE': 0, 'REAL': 0}
                frequency_dictionary[word] = {'FAKE': 0, 'REAL': 0}
            
            # Increment the count for the specific label ('FAKE' or 'REAL')
            frequency_dictionary[word][label] += 1
    
    # Return the frequency dictionary
    return frequency_dictionary

# Create the frequency dictionary using the training data and labels
frequency_dictionary = create_frequency_dictionary(X_train, y_train)

# Extract the vocabulary (set of unique words) from the frequency dictionary
vocabulary = set(frequency_dictionary.keys())

````

````python
# Step 4: Convert data to bag-of-words vectors using closed vocabulary
# The purpose of this method is to create a feature vector for each text based on the vocabulary. 
# The vector indicates how many times each word in the vocabulary appears in the given text.

def extract_features(text, vocabulary, frequency_dictionary):
    # Tokenize the input text into individual words
    words = tokenize(text)
    
    # Create an empty dictionary to store the feature vector (word counts)
    features = {}
    
    # For each word in the vocabulary, count its occurrences in the current text
    # The method checks each word in the vocabulary and records how many times it appears in the text
    for word in vocabulary:
        features[word] = words.count(word)
    
    return features  # Return the feature vector (word counts)

# Generate the feature vectors for the training data
# The method `extract_features` is applied to each text in the training set (X_train).
# It returns a list of feature vectors representing each text's word counts based on the vocabulary.
train_features = [extract_features(text, vocabulary, frequency_dictionary) for text in X_train]

# Generate the feature vectors for the testing data
# Similarly, the method `extract_features` is applied to each text in the testing set (X_test).
# It returns a list of feature vectors for the test set as well.
test_features = [extract_features(text, vocabulary, frequency_dictionary) for text in X_test]
````

````python
# Step 5: Train a Naive Bayes Classifier using NLTK
from nltk.classify import NaiveBayesClassifier


# Create a training set by combining features with their labels
# train_features contains the feature dictionaries for each example
# y_train contains the class labels (e.g. categories)
train_set = list(zip(train_features, y_train))

# Train the Naive Bayes Classifier on the training set
# This method takes the training data and builds a model
# that can be used to predict the category of new text
classifier = NaiveBayesClassifier.train(train_set)
````

````python
# Step 6: Evaluate the classifier using accuracy and confusion matrix
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Get predictions
predictions = [classifier.classify(feat) for feat in test_features]

# Compute accuracy
accuracy = accuracy_score(y_test, predictions)
print(f"Accuracy: {accuracy:.4f}")

# Create confusion matrix
cm = confusion_matrix(y_test, predictions, labels=['FAKE', 'REAL'])
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['FAKE', 'REAL'])
disp.plot()
plt.show()
````

````python
# Step 6: Evaluate the classifier using accuracy and confusion matrix

# Import methods to measure model performance and to display results clearly
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Make predictions for the test dataset
# The classifier will use the test features to guess the label (FAKE or REAL) for each item
predicted_labels = [classifier.classify(feature_set) for feature_set in test_features]

# Calculate how many of the predicted labels match the actual labels
# This gives us an accuracy score between 0 and 1
classification_accuracy = accuracy_score(y_test, predicted_labels)
print(f"Accuracy: {classification_accuracy:.4f}")

# Create the confusion matrix to see how often the model predicted each class correctly
# It shows correct predictions on the diagonal and errors off the diagonal
confusion_matrix_result = confusion_matrix(y_test, predicted_labels, labels=['FAKE', 'REAL'])

# Display the confusion matrix in a more visual and easy-to-read way
# This makes it easier to spot where the classifier is performing well or struggling
confusion_matrix_display = ConfusionMatrixDisplay(confusion_matrix=confusion_matrix_result, display_labels=['FAKE', 'REAL'])
confusion_matrix_display.plot()
plt.show()
````
